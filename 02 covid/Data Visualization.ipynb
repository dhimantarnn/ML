# Data Analyzation

```python
import pandas as pd
```

```python
df = pd.read_csv('covid_19_india.csv')
```

```python
df.head()
```

```python
df = df[['Date', 'State/UnionTerritory', 'Cured', 'Deaths', 'Confirmed']]
```

```python
df.columns = ['date', 'state', 'cured', 'deaths', 'confirmed']
```

```python
df = df.dropna()
```

```python
max = df.sort_values(by= "confirmed", ascending=False)
```

```python
max
```

```python
top = max[:10]
```

```python
top
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x='state', y='confirmed', data= top, hue='state')
plt.show()
```

```python
max_deaths = df.sort_values(by="deaths", ascending= False)
max_deaths
```

```python
top_deaths = max_deaths[:10]
```

```python
sns.set({'figure.figsize':(15,10)})
sns.barplot(x= 'state', y= 'deaths', hue= 'state', data= top_deaths)
plt.show()
```

```python
max_cured = df.sort_values(by= 'cured', ascending= False)
top_cured = max_cured[:10]
top_cured
```

```python
sns.set({'figure.figsize': (15,10)})
sns.barplot(x= 'state', y= 'cured', hue='state', data= top_cured)
plt.show()
```

```python
maha = df[df.state == 'Maharashtra']
maha
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x= 'date', y='confirmed', data= maha)
plt.show()
```

```python
sns.set({'figure.figsize':(15,10)})
sns.lineplot(x= 'date', y='deaths', data= maha)
plt.show()
```
